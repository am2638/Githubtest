(2) Suppose $x$ is a variable from normal distribution, whose mean is 3 and variance is 25. What is the probability of $-10 < x < 16$.

Make a plot like this:
![normal](http://nycdatascience.com/slides/BOOTCAMP2/content/week2/pic/pro.png)

Replace the "?" with right answer in the plot.

```python
### write your code here
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
var = 25
mu, sigma = 3,5
x = np.linspace(-10,16,1000)
y = stats.norm.pdf(x, mu, sigma)
plt.figure(figsize=(8,6))
plt.plot(x,y)
plt.fill(x,y, alpha = 0.3)

plt.title('Normal Distribution: $N($%d, $%d)$'%(mu, var))
plt.text(-1.9, 0.01, '$p(-10 < x < 16)=%.4f$'%(1 - 2*stats.norm.cdf(3,5)), size = 15)

plt.show()
```
